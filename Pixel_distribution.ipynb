```python
from pathlib import Path
import os
```

```python
image_path = "/home/cai/anlsts_exl_isc_minerva_Poc_2c0551ba_Sol_21/Sitaram/Image_Metadata_research/image_input/Invoice_4.jpg"
base_name = os.path.basename(image_path)
base_name = base_name[:-4]
base_name
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in BGR format
img_bgr = cv2.imread(image_path)

# Convert to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Extract individual channels
r = img_rgb[:, :, 0].flatten()
g = img_rgb[:, :, 1].flatten()
b = img_rgb[:, :, 2].flatten()

# Compute histograms
hist_r, _ = np.histogram(r, bins=256, range=(0, 256))
hist_g, _ = np.histogram(g, bins=256, range=(0, 256))
hist_b, _ = np.histogram(b, bins=256, range=(0, 256))

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Red channel
axs[0].plot(hist_r, color='red')
axs[0].set_title('Red Channel Intensity')
axs[0].set_xlabel('Pixel Intensity')
axs[0].set_ylabel('Frequency')
axs[0].grid(True)

# Green channel
axs[1].plot(hist_g, color='green')
axs[1].set_title('Green Channel Intensity')
axs[1].set_xlabel('Pixel Intensity')
axs[1].set_ylabel('Frequency')
axs[1].grid(True)

# Blue channel
axs[2].plot(hist_b, color='blue')
axs[2].set_title('Blue Channel Intensity')
axs[2].set_xlabel('Pixel Intensity')
axs[2].set_ylabel('Frequency')
axs[2].grid(True)

plt.tight_layout()

output_path = f"RGB_histogram_plot_{base_name}.png"  # Specify the output file path and name
plt.savefig(output_path)
plt.show()
# plt.show()
```

### PIXEL DISTRIBUTION IN GRAYSCALE CHANNEL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Flatten and compute histogram
hist, bins = np.histogram(img.flatten(), bins=256, range=[0, 256])

# Plot as a curve
plt.figure(figsize=(8, 5))
plt.plot(hist, color='black')
plt.title("Grayscale Pixel Intensity Curve")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")
plt.grid(True)


# Save the plot to an image file
output_path = f"histogram_plot_{base_name}.png"  # Specify the output file path and name
plt.savefig(output_path)
plt.show()
```
